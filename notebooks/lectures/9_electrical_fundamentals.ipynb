# Alternating current

```python
import matplotlib
# matplotlib.use("Agg")
import matplotlib.pyplot as plt
import numpy as np

matplotlib.rcParams.update({"font.family": "serif"})

WIDTH, ASPECT = 5.5, 9/16
fig, ax = plt.subplots(figsize=(WIDTH, WIDTH * ASPECT))

# ── Signal ────────────────────────────────────────────────────────────────────
Im    = 1.0
T     = 1.0
w     = 2 * np.pi / T
theta = np.pi / 2          # lag: peak sits at t = -theta/w, left of y-axis

t_peak = -theta / w        # ≈ -0.167

t_start = -0.72
t_end   =  2.25 * T

t      = np.linspace(t_start, t_end, 1000)
i_wave = Im * np.cos(w * t + theta)

ax.plot(t, i_wave, 'k-', linewidth=1.2)

# ── Axes drawn as full arrows ─────────────────────────────────────────────────
for spine in ax.spines.values():
    spine.set_visible(False)
ax.set_xticks([])
ax.set_yticks([])

x_min, x_max = t_start - 0.05, t_end + 0.15
y_min, y_max = -Im * 1.6,       Im * 1.7

ax.set_xlim(x_min, x_max)
ax.set_ylim(y_min, y_max)

_AW = dict(arrowstyle='->', color='k', lw=0.9, mutation_scale=9)
ax.annotate('', xy=(x_max, 0), xytext=(x_min, 0),
            arrowprops=_AW, annotation_clip=False)
ax.annotate('', xy=(0, y_max), xytext=(0, y_min),
            arrowprops=_AW, annotation_clip=False)

ax.text(x_max -0.04, -0.05, r'$\omega t$', ha='left', va='top',    fontsize=11, style='italic')
ax.text(0.03,  y_max-0.1, 'i', ha='left', va='bottom', fontsize=11, style='italic')

# Im tick and label on y-axis
ax.plot([-0.02, 0.02], [Im, Im], 'k-', lw=0.8)
ax.text(0.04, Im-0.04, '$I_m$', ha='left', va='top', fontsize=10)

# ── Shared annotation helpers ─────────────────────────────────────────────────
ann_y = Im + 0.2
cap_h = 0.1
_DA   = dict(arrowstyle='<->', color='k', lw=0.8, mutation_scale=7,shrinkA=0, shrinkB=0)

def span(x0, x1, label, label_x=None):
    ax.annotate('', xy=(x1, ann_y), xytext=(x0, ann_y), arrowprops=_DA)
    for x in (x0, x1):
        ax.plot([x, x], [ann_y - cap_h, ann_y + cap_h], 'k-', lw=0.8)
    lx = (x0 + x1) / 2 if label_x is None else label_x
    ax.text(lx, ann_y + 0.08, label, ha='center', va='bottom', fontsize=10)

ax.annotate('', xy=(t[0], 0), xytext=(t[0], Im), arrowprops=_DA)
ax.text(t[0]-0.04,Im/2, 'Magnitude,\n amplitude', ha='right', va='center', fontsize=10)

# ── θ: y-axis to lagged peak ──────────────────────────────────────────────────
span(t_peak, 0, r'$\theta$')

# ── ωT = 2π: successive peaks to the right of y-axis ─────────────────────────
peak2 = t_peak + T
peak3 = t_peak + 2 * T
span(peak2, peak3, r'$\omega T = 2\pi$')
ax.axhline(xmin=0,xmax=(peak2-t[0])/(t[-1]-t[0]),y=Im,ls="--")

# ── Save ─────────────────────────────────────────────────────────────────────
fig.subplots_adjust(left=0.10, right=0.95, top=0.90, bottom=0.08)
fig.savefig("figure.png", dpi=150)
plt.show()
```
