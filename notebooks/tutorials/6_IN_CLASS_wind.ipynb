```python
import numpy as np
import matplotlib.pyplot as plt

α,Δd = 0.1, 15
af = np.linspace(0,0.5,1000)
front = 27.0/4.0 * af * (1-af)**2
c = np.sqrt((1-2*af)/(1-af))
rear = (1-2*af*(1/(1+2*α*Δd*c))**2 )**3
total = front + rear
idx_max = np.argmax(total)

fig,ax = plt.subplots(figsize=(7,3))
ax.plot(af,front,label='Front',alpha=0.3)
ax.plot(af,rear,label='Rear',alpha=0.3)
ax.plot(af,total,label='Total')
ax.axvline(1/3.0,ls='--',color='red',alpha=0.2,label='Single turbine optimum')
ax.plot([af[idx_max],af[idx_max]],[0,total[idx_max]],ls=':',color='black')

ax.set_ylim(0,2.5)
ax.set_xlabel(r'Front axial induction, $a_f$')
ax.set_ylabel('Normalized power')
ax.set_title(fr'Spacing: {Δd:.1f}D, $a_f^*={af[idx_max]:.3f}$, Power = {total[idx_max]:.2f}')
ax.legend()

af = 1.0/3.0
front = 27.0/4.0 * af * (1-af)**2
c = np.sqrt((1-2*af)/(1-af))
rear = (1-2*af*(1/(1+2*α*Δd*c))**2 )**3
print(f'Self optimized: {front+rear:.2f}')
```
